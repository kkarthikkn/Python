<h2>Select the columns to keep</h2>

```python
col_to_keep=['Age', 'Attrition', 'BusinessTravel', 'DailyRate', 'Department',
       'DistanceFromHome', 'Education', 'EducationField', 
       'EnvironmentSatisfaction', 'Gender', 'HourlyRate',
       'JobLevel', 'JobRole', 'JobSatisfaction',
       'MaritalStatus', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked',
       'Over18', 'OverTime','TotalWorkingYears', 'WorkLifeBalance',
       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager']
```

```python
len(col_to_keep)
```

```python
# update the columns to df
df=df[col_to_keep]
df
```

```python
# check for duplicates
df.isna().sum()     

'''  no duplicates  ''' 
```

----

```python
# renaming column
df.rename(columns={'DistanceFromHome':'DistanceHome'},inplace=True)
```

```python
df.rename(columns={'TotalWorkingYears':'Experience'},inplace=True)
df
```

```python
# renaming multiple columns
df.rename(columns={'DailyRate':'DailyWage','MonthlyRate':'MonthlyWage','HourlyRate':'HourlyWage'},inplace=True)
df
```

```python
# replacing the values in the column
df['BusinessTravel']=df['BusinessTravel'].str.replace('_',' ')
df
```

<h2>Filtering</h2>

```python
# filters the row based on the given condition
df[df['BusinessTravel']=='Travel Frequently']
```

```python
df[df['Age']<=35].sort_values(by='Age',ascending=True)
```

```python
df[(df['Experience']>=10) & (df['MaritalStatus']=='Married')].sort_values(by='Experience',ascending=False)
```

<h2>Formatting Column Values</h2>

```python
df.head()
```

<b>NOTE : </b><i> Check the values before formatting the values

```python
df['Attrition'].value_counts()
```

```python
''' Attrition column '''

# replacing multiple values in single col
df['Attrition'].replace({'Yes':'Y','No':'N'},inplace=True) 
df.head()
```

```python
df['Gender'].value_counts()
```

```python
''' Gender column '''

df['Gender'].replace({'Male':'M','Female':'F'},inplace=True)
df.head()
```

```python
df['OverTime'].replace({'Yes':'Y','No':'N'},inplace=True)
df.head()
```

```python
# replacing the values with 0's & 1's
df['Attrition']=df['Attrition'].apply(lambda x: 1 if x=="Y" else 0)
```
